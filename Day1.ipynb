interpolate()関数を使うと，欠損値の補完を行ってくれる．しかし，Nanが挟まっている両隣の平均値を埋めるので，今回はこの手法はふさわしくない．

では今回，性別ごとの平均値で埋めてみよう

```python
female_age_mean = round(df[df.Sex == 'female']['Age'].mean())
male_age_mean = round(df[df.Sex == 'male']['Age'].mean())

print('女性の年齢の平均値は{0}歳，男性の年齢の平均値は{1}歳です．この平均値を補完します．'.format(female_age_mean, male_age_mean))

df[df.PassengerId == 20][['PassengerId', 'Name', 'Sex', 'Age']]
```

```python
#df = pd.read_csv('train.csv')
dff = df[df.Sex=='female'].fillna({'Age': female_age_mean})
dfm = df[df.Sex=='male'].fillna({'Age': male_age_mean})

df = dff.append(dfm)
```

```python
df[df.PassengerId==6][['PassengerId', 'Name', 'Sex', 'Age']]
```

fillnaで欠損値の補完

```python
def classification_age(age):
    if age <= 19:
        return '1'
    elif age <= 34:
        return '2'
    elif age <= 49:
        return '3'
    elif age >= 50:
        return '4'    
    else:
        return '0'
    
df['AgeClass'] = df.Age.map(classification_age)   #mapは各要素に関数を全て適用するやつ
df.head()
```
